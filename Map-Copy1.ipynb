```python
#Import all necessary packages
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import json
import operator
import numpy as np
import ast
import folium
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df=pd.read_csv("for_map_df.csv")
df.Ingredients = df.Ingredients.apply(ast.literal_eval)
```

```python
############################### dictionary of ingredients use of each cuisine 
d={}
diff=[]
aver=[]
l = list(df['Cuisine'].unique())
for cuis in l:
    ingr_dict_cuis={}
    for ing in df[df['Cuisine']==cuis]['Ingredients']:
        for i in ing:
            if i in ingr_dict_cuis:
                ingr_dict_cuis[i] = ingr_dict_cuis[i]+1
            else:
                ingr_dict_cuis[i] = 1
    d[cuis] = ingr_dict_cuis
    
############################### list of difficulty of each cuisine (from 0 to 1, 1 being very difficult, 0 being very easy)

for cuis in l:
    diff.append(1-sum(df[df['Cuisine']==cuis]['Easy'].eq('Yes').mul(1).values)/len(df[df['Cuisine']==cuis]))
            
############################## list dataframes of ingredients and their count of each cuisine
my_data=[]

for i in range(len(l)):
    a=pd.DataFrame.from_dict(d.get(l[i]),orient='index',columns=['Count'])
    a['Proportion']=a.Count/len(a)
    my_data.append(a)

# top 5 ingredients used in each cuisine
cc=[list(i.sort_values(by=['Count'],ascending=False).head(5).index) for i in my_data]

for i in range(len(my_data)):
    my_data[i]['Ingredient'] = my_data[i].index
my_data = [x.reset_index(drop=True) for x in my_data]

########################### dataframe of cuisine, difficulty and average  number of Ingredeines
for i in range(15):
    aver.append(sum(my_data[i]['Count'].values)/len(df[df['Cuisine']==l[i]]))

diff_av = pd.DataFrame(list(zip(l, diff,aver)),
               columns =['Cuisine', 'Difficulty', 'Average Number of Ingredients'])
```
